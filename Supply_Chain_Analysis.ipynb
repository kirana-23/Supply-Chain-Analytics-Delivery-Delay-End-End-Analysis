```python
df.drop(columns=['Customer Email', 'Customer Fname', 'Customer Lname', 'Customer Password', 'Customer Street',
    'Benefit per order', 'Sales per customer', 'Order Item Product Price', 'Product Card Id','Product Category Id',
    'Order Customer Id','Order Zipcode', 'Customer Zipcode', 'Product Status', 'Product Description', 'Product Image',
    'Latitude', 'Longitude','Order Id', 'Order Item Id','Department Id', 'Category Id'],inplace=True)
```

```python
df["Delivery Status"].unique()
```

```python
df["is_late"]=df["Delivery Status"].apply(lambda x: 1 if "Late delivery" in str(x) else 0)
```

```python
df["is_late"].head(10)
```

```python
df["Delay_ship_days"]=df["Days for shipment (scheduled)"]-df["Days for shipping (real)"]
df["Delay_ship_days"]
```

```python
status_count=df["Delivery Status"].value_counts()
sns.barplot(x=status_count.index,y=status_count.values,palette="Set2")
plt.xlabel("Delivery status")
plt.ylabel("Number of orders")
plt.title("Delivery status distribution")
plt.tight_layout()
plt.show()
```

```python
region_wise=df.groupby("Order Region")["is_late"].mean().sort_values(ascending=False)*100
plt.figure(figsize=(12,6))
sns.barplot(x=region_wise.index,y=region_wise.values,palette="Blues_r")
plt.xticks(rotation=45,ha="right")
plt.xlabel("Region")
plt.ylabel("Late percentage")
plt.title("Region by late percentage")
plt.tight_layout()
plt.show()
```

```python
customer_seg=df.groupby("Customer Segment")["is_late"].mean().sort_values(ascending=False)*100
sns.barplot(x=customer_seg.index,y=customer_seg.values,palette="Reds_r")
plt.xlabel("Customer segmentation")
plt.ylabel("Late delivery percentage")
plt.title("Customer segmentation by Late delivery percentage")
plt.show()
```

```python
df["Category Name"].unique()
```

```python
category_wise=df.groupby("Category Name")["is_late"].mean().sort_values(ascending=False)*100
plt.figure(figsize=(24,6))
sns.barplot(x=category_wise.index,y=category_wise.values,palette="Greens_r")
plt.xticks(rotation=45,ha="right")
plt.xlabel("Category Name")
plt.ylabel("Late delivery percentage")
plt.title("Category by Late delivery percentage")
plt.show()
```

```python
ship_wise=df.groupby("Shipping Mode")["is_late"].sum().sort_values(ascending=False)
sns.barplot(x=ship_wise.index,y=ship_wise.values,palette="magma")
plt.xlabel("Shipping mode")
plt.ylabel("Late delivery")
plt.title("Late delivery by shipping mode")
plt.show()
    
```

```python
data=df.groupby("Type")["is_late"].sum().sort_values(ascending=False)
sns.barplot(x=data.index,y=data.values,palette="Purples_r")
plt.xlabel("Type of payment")
plt.ylabel("Late delivery")
plt.title("Late delivery by payment type")
```

```python
#logistic performance
plt.figure(figsize=(30,6))
sns.barplot(x=df["Category Name"],y=df["Sales"])
plt.xticks(rotation=45,ha="right")
plt.title("Sales by Category")
plt.show()
```

```python
df["order date (DateOrders)"]=pd.to_datetime(df["order date (DateOrders)"])
```

```python
df["Order Year"]=df["order date (DateOrders)"].dt.year
```

```python
df["Order Month"]=df["order date (DateOrders)"].dt.month
```

```python
data=df.groupby(["Order Year","Order Month"])["Sales"].count().reset_index()
data.columns=["Year","Month","Count"]
data["Date"]=pd.to_datetime(data[["Year","Month"]].assign(day=1))
plt.figure(figsize=(24,6))
plt.plot(data["Date"],data["Count"],marker="o",linewidth=3,color="maroon")
plt.xlabel("Year-Month")
plt.ylabel("Order volume")
plt.title("Monthly Order volume trend")
plt.show()
```

```python
data=df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=data.values,y=data.index,palette="Greens_r")
plt.xlabel("Product sales")
plt.ylabel("Product Name")
plt.title("Top 10 Porducts by sales")
```

```python
data=df.groupby("Category Name")["Sales"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=data.values,y=data.index,palette="viridis")
plt.xlabel("Category sales")
plt.ylabel("Category Name")
plt.title("Top 10 Category by sales")
```

```python
data=df.groupby("Category Name")['Order Profit Per Order'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=data.values,y=data.index,palette="grey_r")
plt.xlabel("Average profit")
plt.ylabel("Category Name")
plt.title("Top 10 Category by Average profit")
```

```python
data=df.groupby("Category Name")["Order Profit Per Order"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=data.values,y=data.index,palette="magma")
plt.xlabel("Total Profit")
plt.ylabel("Category")
plt.title("Top 10 Category by profit")
```

```python
data=df.groupby("Order City")["Sales"].count().sort_values(ascending=False).head(10)
sns.barplot(x=data.index,y=data.values,palette="grey_r")
plt.xticks(rotation=45,ha="right")
plt.xlabel("City")
plt.ylabel("order frequency")
plt.title("Top 10 cities by Order frequency")
```
